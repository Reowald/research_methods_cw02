```python
!pip install newspaper3k
!pip install regex

import newspaper
from newspaper import Article
import regex as re

def _write_article_to_text_file(article, folder_path, source_name, i):
    try:
        print(article.url)
        # download from website
        article.download()
        #read article to be able to get text
        article.parse()
        #get only text from article
        text = article.text


    #create a text file in output folder path with name source and number of article
    #put a try except so that any articles with weird characters get skipped

        with open(f'{folder_path}/{source_name}_{i}.txt', 'w') as file:
            # Write the text content to the file
            file.write(text)
    except:
        pass

def news_article_scraper(domain:str, folder_path:str, source_name:str):
    """
    Parameters:
    ----------
    domain: main url of news website
    folder_path: path to folder you want to download the articles to
    source_name: name of news website to appear at start of file name
    
    """
    #build "newspaper" by finding all urls from domain
    news_site = newspaper.build(domain, memoize_articles=False)

    #get list of articles within site - urls
    articles = news_site.articles

    #loop through list of articles
    for i, article in enumerate(articles):
        
        #go through each source and check that the articles actually contain news

        #starting with the mirror
        if source_name == 'mirror':
            #check that the url ends with a number, and doesn't contain the word promotion
            if re.search(r'[0-9]+$', article.url) is not None and not re.match(r'.+promotion.+',article.url) and not re.match(r'.+shopping-deals.+',article.url):
                _write_article_to_text_file(article, folder_path, source_name, i)

        elif source_name == 'dailymail':
            #if re.search(r'[0-9]+$', article.url) is not None:
            _write_article_to_text_file(article, folder_path, source_name, i)

        elif source_name == 'thetimes':
            #if re.search(r'[0-9]+$', article.url) is not None:
            _write_article_to_text_file(article, folder_path, source_name, i)

        elif source_name == 'theguardian':
            #if re.search(r'[0-9]+$', article.url) is not None:
            _write_article_to_text_file(article, folder_path, source_name, i)

        elif source_name == 'thesun':
            #if re.search(r'[0-9]+$', article.url) is not None:
            _write_article_to_text_file(article, folder_path, source_name, i)

        elif source_name == 'thedailystar':
            #if re.search(r'[0-9]+$', article.url) is not None:
            _write_article_to_text_file(article, folder_path, source_name, i)

        elif source_name == 'independent':
            #if re.search(r'[0-9]+$', article.url) is not None:
            _write_article_to_text_file(article, folder_path, source_name, i)

        elif source_name == 'financial times':
        #if re.search(r'[0-9]+$', article.url) is not None:
            _write_article_to_text_file(article, folder_path, source_name, i)


news_article_scraper(
    domain = 'https://www.ft.com/',
    folder_path = r'/Users/carlpo/Documents/News Articles',
    source_name = 'independent'
)
```
